```python
from matplotlib import pyplot as plt
import numpy  as np 

sns.set()

x=np.linspace(1,200,20)
y=np.sin(x)


plt.plot(x,y)
```

```python
sns.set_style("dark")

# white dark whitegrid darkgrid ticks 

x=np.linspace(1,200,20)
y=np.sin(x)


plt.plot(x,y)
```

```python
pattern=sns.color_palette("viridis")
sns.palplot(pattern)
```

```python
flower_data=sns.load_dataset('iris')
flower_data.head(3)
sns.pairplot(flower_data,hue='species')
```

```python
tdata=sns.load_dataset('titanic')
tdata.head(5)


tdata['class'].isnull().sum() # no null data 
sns.barplot(x="class",y="pclass",data=tdata)
# one value is numeric and other value is in category 
tdata.head(5)
```

```python
tdata=sns.load_dataset('titanic')
tdata.head(5)

sns.countplot(x='embark_town',data=tdata,hue='class')
```

```python
tdata=sns.load_dataset('titanic')
tdata.head(5)

sns.violinplot(x='age',y='who',data=tdata,hue='sex')  #Neither the `x` nor `y` variable appears to be numeric.
```

# Conclusion 

Type  1

Type 2 

barplot -  data, x, y 
boxplot  -data, x, y - Large data sets
jointplot  - data, x, y 



Type 3 
stripplot -data, x is category and y is normal
swarmplot -data, x is category and y is normal
