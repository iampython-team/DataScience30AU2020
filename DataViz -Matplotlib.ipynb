```python

year=[1950,1970,1990,2010,2012,2014,2016,2018]
population=[67,89,100,102,110,120,125,130]

plt.plot(year,population,'ro',linewidth=20)


plt.title("Population India (year wise) ")

plt.xlabel("year (Alternate 20 years)")
plt.ylabel("Population (in million years)")
```

```python
plt.bar(year,population,width=1)


plt.title("Population India (year wise) ")

plt.xlabel("year (Alternate 20 years)")
plt.ylabel("Population (in million years)")
```
